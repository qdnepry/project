```python
def show_image(image):
  if len(image.shape) == 3:
    height, width, channals = image.shape
  else :
    height, width, channals = image.shape[0], image.shape[1], 0
  fig = plt.figure()
  ax = fig.add_subplot(1, 1, 1)
  major_ticks_x = np.arange(0, width, 40)
  minor_ticks_x = np.arange(0, width, 10)
  major_ticks_y = np.arange(0, height, 40)
  minor_ticks_y = np.arange(0, height, 10)
  ax.set_xticks(major_ticks_x)
  ax.set_xticks(minor_ticks_x, minor=True)
  ax.set_yticks(major_ticks_y)
  ax.set_yticks(minor_ticks_y, minor=True)
  ax.grid(color='white', which='both')
  ax.grid(which='minor', alpha=0.1)
  ax.grid(which='major', alpha=0.2)
  plt.imshow(image)
```

```python
img = cv.imread('img.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img = cv.resize(img, (400, 400), 0.5, 0.5, interpolation=cv.INTER_LINEAR)
```

```python
show_image(img)
```

```python
width = range(295, 305, 1)
height = range(310, 320, 1)
```

```python
for y in height:
  for x in width:
    color = img[y, x]
    print(f'coord ({x}, {y}), color:{color}')
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
show_image(gray_img)
```

```python
binary_image = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 5, 1)
show_image(binary_image)
```

```python
lower_brown = np.array([117, 102, 27], dtype=np.uint8)
upper_brown = np.array([202, 187, 114], dtype=np.uint8)
mask = cv.inRange(img, lower_brown, upper_brown)
# show_image(mask)
```

```python
show_image(mask)
```
